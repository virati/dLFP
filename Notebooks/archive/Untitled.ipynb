```python
import networkx as nx
import scipy.signal as sig
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def fc(x,t,L):
    return -np.dot(L,x)

f_c = lambda x,t,L: -np.dot(L,x).squeeze()
t = np.arange(0,10,1/100)

N = 8
G = nx.complete_graph(N)
L = np.eye(N)
#L = nx.linalg.laplacianmatrix.laplacian_matrix(G).todense()
print(L)
print(L.shape)
y0 = np.random.uniform(-10,10,size=(N,))
sol = odeint(fc,y0,t,args=(L,)) 
```

```python
print(sol.shape)
plt.plot(sol)
plt.plot(np.mean(sol,axis=1),linestyle='--',linewidth=2,color='black')
plt.text(600,2.5,str(np.mean(sol,axis=1)[-1]))
```
