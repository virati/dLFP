```python
import networkx as nx
import scipy.signal as sig
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def f_c(x,t,params):
    L = params['L']
    x_dot = -np.dot(L,x)
    
    return x_dot

def unity(x,t,params):
    return x

class consensus:
    def __init__(self,N,f=unity):
        self.f_drift = f
        self.N = N
        self.G = nx.complete_graph(N)  
        self.L = np.array(nx.linalg.laplacianmatrix.laplacian_matrix(self.G).todense())
        self.t = np.arange(0,10,1/100)

    def plot_L(self):
        print(L)
        print(L.shape)

    def sim(self,x_0,t):
        
        params = {'L':self.L}
        sol = odeint(self.f_drift,x_0,t,args=(params,))
        return sol
    
    def sim_plot(self,x_0,t=0):
        sol = self.sim(x_0 = x_0,t=self.t)
        print(sol.shape)
        plt.plot(sol)
        plt.plot(np.mean(sol,axis=1),linestyle='--',linewidth=2,color='black')
        plt.text(600,2.5,str(np.mean(sol,axis=1)[-1]))

N = 10
x_0 = np.random.uniform(-10,10,size=(N,))
system = consensus(N=N,f=f_c)
system.sim_plot(x_0=x_0)
```
