# Why I don't filter my timeseries

Filtering timeseries when you're doing frequency-domain analysis introduces many headaches.
That's why I don't filter my timeseries before analysing oscillations.

This notebook will walk you through why

## Setup

```python
import DBSpace as dbo
import DBSpace.dLFP.diff_model as sim
import matplotlib.pyplot as plt
import scipy.signal as sig
import numpy as np

plt.rcParams['figure.figsize'] = [15, 5]
```

## Simulator

```python
diff_run = sim.sim_diff(Ad=2000,wform='moresine4',clock=True,stim_v=6,stim_freq=130)
amp_run = sim.sim_amp(diff_run,family='tanh',noise=1e-6,sig_amp_gain=10,pre_amp_gain=1)
amp_run.simulate(1000,1100)
timeseries = amp_run.sim_output_signal
```

## Analysis
So we've got a timeseries that we want to analyse. Let's first do some typical filtering.

```python
sos_lpf = sig.butter(10,20,fs=422,output='sos')
fsg_sig = sig.sosfilt(sos_lpf,timeseries)

plt.plot(timeseries)
plt.plot(fsg_sig)
```

```python
F,T,SG1 = sig.spectrogram(timeseries,nfft=2**10,window='blackmanharris',nperseg=1024,noverlap=1000,fs=422)
F,T,SG2 = sig.spectrogram(fsg_sig,nfft=2**10,window='blackmanharris',nperseg=1024,noverlap=1000,fs=422)

plt.figure()
plt.subplot(121)
plt.pcolormesh(T,F,(np.log10(SG1)))
plt.subplot(122)
plt.pcolormesh(T,F,np.log10(SG2))
```

```python
F,P1 = sig.welch(timeseries,nfft=2**10,window='hann',nperseg=1024,noverlap=1000,fs=422)
F,P2 = sig.welch(fsg_sig,nfft=2**10,window='hann',nperseg=1024,noverlap=1000,fs=422)

plt.figure()
plt.plot(F,np.log10(P1))
plt.plot(F,np.log10(P2))
```

```python
Fsg,Tsg,SG1 = sig.spectrogram(timeseries,nfft=2**10,window='blackmanharris',nperseg=1024,noverlap=1000,fs=422)
Fsg,Tsg,SG2 = sig.spectrogram(fsg_sig,nfft=2**10,window='blackmanharris',nperseg=1024,noverlap=1000,fs=422)

plt.figure()
#plt.plot(F,np.log10(P1))
plt.plot(Fsg,np.mean(np.log10(SG1),axis=1))
plt.plot(Fsg,np.mean(np.log10(SG2),axis=1))
```

```python
b,a = sig.butter(8,(58,62),btype='bandstop',fs=422,output='ba')
#b,a = sig.butter(5,50,btype='highpass',fs=422,output='ba',analog=False)

w, h = sig.freqz(b, a)
plt.semilogx(w, 20 * np.log10(abs(h)))
plt.title('Butterworth filter frequency response')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
#plt.margins(10, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(100, color='green') # cutoff frequency
plt.show()

notched = sig.lfilter(b,a,timeseries)

plt.plot(timeseries)
plt.plot(notched)
```

```python
F,P1 = sig.welch(timeseries,nfft=2**10,window='hann',nperseg=1024,noverlap=1000,fs=422)
F,P2 = sig.welch(notched,nfft=2**10,window='hann',nperseg=1024,noverlap=1000,fs=422)

plt.figure()
plt.plot(F,np.log10(P1))
plt.plot(F,np.log10(P2))
```
